# **Exploratory Data Analysis**

## **What are the main characteristics that have the most impact on the car price?**

### **Import Data from Module 2**

Import libraries:

```python
import pandas as pd
import numpy as np
```

Load the data and store it in dataframe df:

```python
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

### **Analyzing Individual Feature Patterns Using Visualization**

```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

**How to choose the right visualization method?**

```python
# list the data types for each column
print(df.dtypes)
```

```python
# show data type of column "peak-rpm"
print(df['peak-rpm'].dtypes)
```

```python
# show correlation between variables of type "int64" & "float64"
df.corr()
```

```python
# show correlation between columns
df[['bore','stroke','compression-ratio','horsepower']].corr()
```

#### **Positive Linear Relationship**

```python
# Engine size as potential predictor variable of price
sns.regplot(x="engine-size", y="price", data=df)

# ylim is used to set the origin at (0,)
plt.ylim(0,)
```

```python
# correlation between 'engine-size' & 'price'
df[["engine-size", "price"]].corr()
```

Engine size and price have a strong positive correlation, with a nearly perfect diagonal regression line, suggesting engine size as a reliable predictor of price.

#### **Negative Linear Relationship**

```python
# show scatterplot of 'highway-mpg' & 'price'
sns.regplot(x="highway-mpg", y="price", data=df)
```

```python
# correlation between 'highway-mpg' & 'price'
df[['highway-mpg', 'price']].corr()
```

- Highway mpg and price display an inverse/negative relationship, indicating that as highway mpg increases, the price tends to decrease. 
- This suggests that highway mpg could potentially serve as a predictor of price.

#### **Weak Linear Relationship** 

```python
# Let's see if "peak-rpm" is a predictor variable of "price"
sns.regplot(x="peak-rpm", y="price", data=df)
```

```python
# show correlation between 'peak-rpm' & 'price'
df[['peak-rpm','price']].corr()
```

Peak rpm is not a reliable predictor of price due to the close-to-horizontal regression line, scattered data points, and high variability.

**Show correlation between 'stroke' & 'price'**

```python
df[['stroke','price']].corr()
```

```python
# display regplot for above correlation
sns.regplot(x="stroke", y="price", data=df)
```

There is a weak correlation between the variable 'stroke' and 'price.' as such regression will not work well.

**Show relationship between "body-style" and "price"**

```python
sns.boxplot(x="body-style", y="price", data=df)
```

We see that the distributions of price between the different body-style categories have a significant overlap, so body-style would not be a good predictor of price.

**Show relationship between "engine-location" and "price"**

```python
sns.boxplot(x="engine-location", y="price", data=df)
```

Here we see that the distribution of price between these two engine-location categories, front and rear, are distinct enough to take engine-location as a potential good predictor of price.

**Show relationship between "drive-wheels" and "price"**

```python
sns.boxplot(x="drive-wheels", y="price", data=df)
```

The distribution of price varies across different drive-wheels categories, suggesting that drive-wheels could potentially be a predictor of price.
